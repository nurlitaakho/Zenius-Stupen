```python
df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
                   
df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})
```

```python
#import library and filter warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import pandasql as ps

import warnings
warnings.filterwarnings('ignore')
```

### Create a dataframe (df3) which consists both dataframe inner-joined by column "Key" (5 pts)

```python
#create df3 dataframe which consists of the result of df1 and df2 inner-joined by column "Key"
df3 = pd.merge(df1, df2, on='key', how='inner')
df3
```

### Create a dataframe (df4) which consists of the result of df1 and df2 left-joined by column "Key" (10pts)

```python
#Create df4 dataframe which consists of the result of df1 and df2 left-joined by column "Key"
df4 = pd.merge(df1, df2, on='key', how='left')
df4
```

# Dataframe Processing


Create a dictionary and convert it to a dataframe called "Customer_df" (15 pts)


name=['Anna','Jason','Cindy']

age=['22','23','21']

gender=['F','M','F']




```python
#create customer dictionary
customer_dict = ({
    'name' : ['Anna','Jason','Cindy'],
    'age' : ['22','23','21'],
    'gender' : ['F','M','F']
})
customer_dict
```

```python
#convert to dataframe
Customer_df = pd.DataFrame.from_dict(customer_dict)
Customer_df
```

### Delete the "age" column (5 pts)

```python
#delete age column
Customer_df_without_age = Customer_df.drop('age', axis=1)
Customer_df_without_age
```

### Rename the "name" column into "first name" (10 pts)

```python
#rename name column into first name
Customer_df = Customer_df.rename(columns = {
    'name' : 'first name'
})
Customer_df.head()
```

### Sort the dataframe by age, descending (10 pts)

```python
#sort dataframe descending by age
Customer_df.sort_values('age', ascending=False)
```

Show the total amount of passengers who survived and not survived (10pts)

Show the average fare paid based on gender (10 pts)

```python
#average fare paid based on gender
df.groupby('Sex')['Fare'].mean()
```

Show the total amount of passengers who survived and not survived based on gender (10pts)

```python
#Total amount of passengers who survived and not survived based on gender
df.groupby(['Survived', 'Sex']).agg({'Survived':['count']})
```
