# Basic Data Operation

You sold 20 pieces of electronics at the price of 12000. The customer gets a total discount of 10%.

Below are those variables:

1.   price=12000
2.   quantity_sold=20
3.   discount=10%


Using Python code, write the code the find the total money that the customer to pay. (Points : 15)

```python
price = 12000
quantity_sold = 20
discount = 10/100

#Total uang yang dibayar oleh pelanggan didapat dari harga barang dikali jumlah item yg terjual,
##kemudian hasilnya dikurangi dengan harga barang dikalikan jumlah barang terjual dikalikan total diskonnya
total_money = (price*quantity_sold)-(price*quantity_sold*discount)
total_money
```

```python
# Replace customer gender list value (Points : 20)
# gender 1 => Male
# gender 0 => Female 

##Bisa menggunakan looping
for i in range (len(customer['gender'])):
  ##Jika value pada key bernilai 1, maka replace menjadi 'Male'
  if customer['gender'][i] == 1 :
    customer['gender'][i] = 'Male'
  ##Kemudian, jika value pada key bernilai 0, maka replace menjadi 'Female'
  elif customer['gender'][i] == 0 :
    customer['gender'][i] = 'Female'

customer
```
