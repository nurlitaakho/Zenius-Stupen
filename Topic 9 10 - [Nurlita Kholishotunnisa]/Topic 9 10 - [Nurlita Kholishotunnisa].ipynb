### Task 2: Plot boxplots for BMI, BP and Y depending on gender

```python
#import package pyplot from matplotlib
import matplotlib.pyplot as plt
```

```python
#import package seaborn
import seaborn as sns
```

```python
#Boxplot for BMI, BP, and Y grouped by SEX
f,ax = plt.subplots(1,3, figsize=(10,4))

#BMI
g = sns.boxplot(data=df,x='SEX', y='BMI', showmeans=True, ax=ax[0])
ax[0].set_title('BMI grouped by SEX')
ax[0].set_xlabel('Sex')

#BP
g = sns.boxplot(data=df,x='SEX', y='BP', showmeans=True, ax=ax[1])
ax[1].set_title('BP grouped by SEX')
ax[1].set_xlabel('Sex')

#Y
g = sns.boxplot(data=df,x='SEX', y='Y', showmeans=True, ax=ax[2])
ax[2].set_title('Y grouped by SEX')
ax[2].set_xlabel('Sex')
```

### Task 3: What is the the distribution of Age, Sex, BMI and Y variables?

```python
#Distribusi untuk variabel Age, Sex, BMI, dan Y
f,ax = plt.subplots(2,2,figsize=(12,7))

#Age
g = sns.histplot(df['AGE'],kde=True, ax=ax[0,0])
ax[0,0].set_title('Age Distribution of Diabetes Patients')
ax[0,0].set_xlabel('')

#Sex
g = sns.histplot(df['SEX'], ax=ax[0,1])
ax[0,1].set_title('Sex Distribution of Diabetes Patients')
ax[0,1].set_xlabel('')

#BMI
g = sns.histplot(df['BMI'],kde=True, ax=ax[1,0])
ax[1,0].set_title('BMI Distribution of Diabetes Patients')
ax[1,0].set_xlabel('')

#Y
g = sns.histplot(df['Y'],kde=True, ax=ax[1,1])
ax[1,1].set_title('Y Distribution of Diabetes Patients')
ax[1,1].set_xlabel('')
```

Berdasarkan output terlihat bahwa:
- untuk variabel Age, sebaran datanya cenderung miring ke kiri (left skewed). Hal ini menandakan bahwa nilai mean < median < modus.
- untuk variabel BMI, sebaran datanya cenderung miring ke kanan (right skewed). Hal ini menandakan bahwa nilai mean > median > modus.
- untuk variabel Y, sebaran datanya juga cenderung miring ke kanan (right skewed). Hal ini menandakan bahwa nilai mean > median > modus.
